```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
n, bins, patches = plt.hist(np.random.random(100000), bins=100)
plt.plot(bins[:-1], n / np.sum(n))
plt.xlabel('sampled value')
plt.ylabel('probability')
plt.title('uniform sampling')
plt.show()
```
